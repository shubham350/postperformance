```python
import pandas as pd

df = pd.read_csv('dataset_Facebook.csv')
df.head()
```

Category : This has Categoriacal data. Considering that it represents category for the post it is nominal.

Post Month, Weekday and Hour : These variables give the time the content was put on Facebook.

Paid : This is a binary data representing 0 when the post was not paid for promotion and 1 when it was paid for promotion.

Lifetime Post Total Reach : This represents the number of people who saw this post.

Lifetime Post Total Impressions : This represents the number of times the post was visible on the screen. The same post may appear more than once on someone's screen.

Lifetime Engaged Users : This reperesnts the clicks anywhere on the post by unique users.This is higher than Lifetime Post Consumers.

Lifetime Post Consumers : This is the unique number of users that clicked anywhere on the post. As this number is less than the previous variable surely it does not account to some clicks. ex. reply to a comment.

Lifetime Post Consumptions : This is the amount of clicks. One consumer might click more than once on the post. S/he may like and comment, etc.

Lifetime Post Impressions by people who have liked your Page : Number of times the post was visible on the screen of people who have liked the post.

Lifetime Post reach by people who like your Page : Unique users who saw the post and have liked the page.

Lifetime People who have liked your Page and engaged with your post : Engaged users who have liked the page.




Some questions that we might ask while understanding the factors that affect the performance or utility of the post :

1. What is our target, whether reaching more people, getting more consumptions/interactions or getting better response form the people it reaches ?
2. What purpose do photos and status serve?
3. Which category performs best?
4. What time is the most appropriate for posting content?
5. Whether paying for a post pays off ?

In this dataset the variables type, category, time of posting and payment for promotion are regressor variables. Rest of the variables are performance measures.




Lifetime post total reach gives an idea of the number of people who saw the post.

Lifetime post consumers and lifetime post engagers give an idea of how many people have engaged in the content. Compared to lifetime post engagers lifetime post consumers will serve our purpose better. A comment thread that is irrelevant to the content will also get counted as engagement. However post consumers will give an idea of relevant clicks to the content.

Post consumers - Total reach gives an idea of the people who saw the post but did not engage.

Post consumers/ Post total reach : This value will give an idea of how effective the post was. A value closer to 1 means effective and one closer to 0 will mean not very effective.

Total page likes gives the number of people associated with your content and want to see more of it.

Lifetime Post reach by people who like your Page will give the number of people to whom it reached.

Lifetime Post reach by people who like your Page / Total page likes will give an idea of how much content has reached the people who like the page. This will have value between 0 to 1

Lifetime Post reach by people who like your Page / Lifetime People who have liked your Page and engaged with your post will give an idea of how satisfied the people associated with your page are. This will have a value between 0 to 1.

```python
#adding new features as response variables

df1 = df.loc[:,'Type': 'Paid']
df1['performance'] = df['Lifetime Post Consumers']/df['Lifetime Post Total Reach']
df1['satisfaction'] = df['Lifetime People who have liked your Page and engaged with your post']/df['Lifetime Post reach by people who like your Page']
```

```python
#checking if the new featues are updated.

df1.head()
```

```python
#checking for missing values

df1.describe()
```

```python
#removing the missing data

df1.dropna(axis='index',how='any',inplace =True)
df1.shape
```

Type, Category, Post month, Post weekday, Post hour and Paid are all categorical variables. Moreover the variables are nominal. Hence we need to preproccess the data. Supposing that the company posts content every month, then we would like to analyse the appropriate time and weekday for performance of the post. So we can drop the post month column. For nominal data, One Hot Encoding is a method for preprocessing so that we can use the data for fitting models. 

```python
#removing the month column

df1 = df1.drop('Post Month',axis = 1)
```

```python
#fixing the regressor and response variables

X = df1.loc[:,'Type':'Paid']
Y = df1.loc[:,'performance':'satisfaction']
```

```python
#dividing the data into train and validation sets. 

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state = 0)
```

```python
#Encoding the nominal data for fitting it to a decision tree. Usual algorithm works for ordinal data. 

from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(handle_unknown='ignore', sparse=False)

#using encoding on the data
EncX_train = pd.DataFrame(enc.fit_transform(X_train))
EncX_test = pd.DataFrame(enc.transform(X_test))

#putting back the index as encoding removes themm
EncX_train.index = X_train.index
EncX_test.index = X_test.index
```

```python
#fitting regression tree with different depths to adjust overfitting and underfitting

from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error

for i in range(2,16,2):
    model = DecisionTreeRegressor(min_samples_leaf=i, random_state=0)
    model.fit(EncX_train,Y_train)
    pred = model.predict(EncX_test)
    mae = mean_absolute_error(pred,Y_test)
    print(i,'model has error', mae )
```

Model with minimum leaves 10 is giving the best results. We can use that to predict how a new post might work. 

```python
#fitting the model on the entire data and then check which posts work the best. 

EncX = pd.DataFrame(enc.fit_transform(X))

EncX.index = X.index

modelf = DecisionTreeRegressor(min_samples_leaf=10, random_state=0)
modelf.fit(EncX,Y)
Ypred = pd.DataFrame(modelf.predict(EncX))
```
